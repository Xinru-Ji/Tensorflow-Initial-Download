```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('./Income1.csv')
```

```python
x = data.Education
y = data.Income
```

```python
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(1,input_shape = (1,)))
```

```python
# 这里的layers是加层，Dense就是模型f(x) = ax + b
# 这里的Dense(1, input_shape=(1,))第一个1是输出的维数，
# 输出的数据维数还是1，输入数据的形状还是一维的，只有
# 一个变量，所以用逗号表示元组
```

```python
# summary反应model整体形状
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam',
             loss='mse' )
```

```python
history = model.fit(x,y,epochs = 5000)
```
